```python
## Regression Analysis with Wins and Goals with mathplot

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset into a Pandas DataFrame
df = pd.read_csv ('C:/Users/User/Dropbox/2023Codes/SQL/laliga21-22.csv')

# Create a scatter plot of Wins vs. Goals Scored
plt.scatter(df['W'], df['GF'])
plt.xlabel('Wins')
plt.ylabel('Goals Scored')
plt.show()

# Perform the regression analysis
X = df['W']
Y = df['GF']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()

# Print the model summary
print(model.summary())
```

Dep. Variable: The dependent variable in the regression analysis. In this case, it's GF (goals scored).
Model: The type of regression model used for the analysis (in this case, it's Ordinary Least Squares or OLS).
R-squared: The coefficient of determination, which measures how well the regression line fits the data. It ranges from 0 to 1, with higher values indicating a better fit. In this case, R-squared is 0.751, which means that 75.1% of the variance in goals scored can be explained by wins.
Adj. R-squared: The adjusted R-squared value that takes into account the number of variables in the model. It's a modified version of R-squared that penalizes the addition of unnecessary variables to the model. In this case, Adj. R-squared is 0.736.
Method: The method used for the regression analysis. In this case, it's Least Squares.
F-statistic: The F-statistic is a measure of how well the regression model as a whole fits the data. It's a ratio of the variance explained by the model to the variance not explained by the model. The higher the F-statistic, the better the model fits the data. In this case, the F-statistic is 51.31, which is quite high.
Prob (F-statistic): The p-value associated with the F-statistic. It measures the probability that the relationship between the dependent variable and the independent variable is due to chance. In this case, the p-value is very low (1.59e-06), which means that the relationship between goals scored and wins is significant.
coef: The coefficients of the regression model. In this case, there are two coefficients: one for the intercept (const) and one for the independent variable (W). The coefficient for W is 2.1909, which means that for every additional win, a team is expected to score 2.1909 more goals.
std err: The standard error of the coefficient estimate. It measures the average amount that the coefficient estimate varies from the true value.
t: The t-value is a measure of how many standard errors the coefficient estimate is from zero. It's a way of testing the hypothesis that the coefficient is equal to zero (i.e., there is no relationship between the independent variable and the dependent variable). In this case, the t-value for W is 7.163, which is quite high, and the p-value is very low, which means that the relationship between wins and goals scored is significant.
P>|t|: The p-value associated with the t-value. It measures the probability that the relationship between the dependent variable and the independent variable is due to chance.

```python
## Random Forest regression model to predict thee league standings of the next 5 years
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset into a pandas dataframe
df = pd.read_csv('C:/Users/User/Dropbox/2023Codes/SQL/laliga21-22.csv')

# Preprocess the data

# Split the data into training and testing sets
train = df[df['Year'] < 2020]
test = df[df['Year'] == 2020]

# Define the features and target variables
features = ['MP', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 'Pts/MP', 'xG', 'xGA', 'xGD', 'xGD/90', 'Attendance percent', 'gsts', 'NDF', 'age', 'poss', 'NPG', 'YC', 'RC', 'GP90', 'XG90', 'XA90', 'Save per', 'Pen save per', 'shots per', 'pass completion', 'prog dist', 'fouls', 'fouls drawn', 'offsides', 'aerial per']
target = 'Pts'

# Create the random forest regression model and fit it to the training data
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(train[features], train[target])

# Evaluate the model's performance on the test data
predictions = rf.predict(test[features])
mse = mean_squared_error(test[target], predictions)
r2 = r2_score(test[target], predictions)

print(f'Mean Squared Error: {mse:.4f}')
print(f'R-squared: {r2:.4f}')

# Make predictions for the next 5 years and display the predicted standings
future_years = pd.DataFrame({'MP': range(38, 193, 38)})
future_predictions = rf.predict(future_years[features])
future_standings = pd.DataFrame({'MP': range(38, 193, 38), 'Predicted Points': future_predictions})
print(future_standings.head())
```

```python
## Predicting the goal count of a team using random forest algorithm

import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Load the data
data = pd.read_csv('C:/Users/User/Dropbox/2023Codes/SQL/laliga21-22.csv')

# Define the features and target variable
X = data.drop(['Squad', 'GF'], axis=1)
y = data['GF']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create the model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the training data
model.fit(X_train, y_train)

# Predict on the test data
y_pred = model.predict(X_test)

# Calculate the accuracy of the model
accuracy = r2_score(y_test, y_pred)

# Print the predicted goal count and accuracy of the model
print("Predicted goal count:", y_pred[:5])
print("Accuracy:", accuracy)
```

```python
# Team Win-Loss Record

import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Load the data
data = pd.read_csv('C:/Users/User/Dropbox/2023Codes/SQL/laliga21-22.csv')

# Drop unnecessary columns
data = data.drop(['Unnamed: 0', 'Squad', 'Pts/MP', 'xGD/90'], axis=1)

# Define target variable
y = data['W']

# Define predictor variables
X = data.drop(['W'], axis=1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the random forest regression model
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model to the training data
rf.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = rf.predict(X_test)

# Calculate the accuracy of the model
accuracy = r2_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
# Visualize the results

import matplotlib.pyplot as plt

# Create scatter plot
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Win-Loss Record")
plt.ylabel("Predicted Win-Loss Record")
plt.title("Random Forest Regression Results")
plt.show()
```
